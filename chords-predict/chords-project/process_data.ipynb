```python
import os
import csv
import shutil
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_labels = pd.read_csv('labels.csv')
df_labels['chord'] = df_labels['chord'].str.replace('/', '_')
labels = df_labels['chord']
labels = pd.DataFrame(labels).drop_duplicates()['chord'].to_list()
```

```python
df_labels
```

```python
source_folder = 'imagens_audio/melspectrogram/'
```

```python
for index, row in df_labels.iterrows():
    image_folder = f'base_data/melspectrogram/raw_data/{row["chord"]}'
    if not os.path.exists(image_folder):
        os.makedirs(image_folder)
    file_name = f"{os.path.splitext(row['file_name'])[0]}.png"
    
    source_path = os.path.join(source_folder, file_name)
    destination_path = os.path.join(image_folder, file_name)
    
    shutil.copy2(source_path, destination_path)
```

```python
original_source = 'base_data/melspectrogram/raw_data'
```

```python
for i in range(len(labels)):
    
    # Extrai o nome de cada classe
    class1 = '/' + labels[i]
    
    # Cria as pastas
    os.makedirs('base_data/melspectrogram/model_data/treino/' + labels[i])
    os.makedirs('base_data/melspectrogram/model_data/val/' + labels[i])
    os.makedirs('base_data/melspectrogram/model_data/teste/' + labels[i])
```

### Cópia das Imagens Para as Pastas

```python
# Loop
for k in range(len(labels)):
    
    # Extrai um nome de classe
    nome_classe = labels[k]
    
    # Define a fonte
    src = original_source + "/" + nome_classe 
    
    # Mostra qual classe estamos processando
    print("\nClasse:", labels[k])
    
    # Lista o conteúdo da pasta
    allFileNames = os.listdir(src)
    
    # "Embaralha" os dados
    np.random.shuffle(allFileNames)
    
    # Divisão = 70% treino, 15% teste, 15% validação 
    train_FileNames, val_FileNames, test_FileNames = np.split(np.array(allFileNames),
                                                          [int(len(allFileNames)*0.7), int(len(allFileNames)*0.85)])


    # Nome dos arquivos
    train_FileNames = [src + '/' + name for name in train_FileNames.tolist()]
    val_FileNames = [src + '/' + name for name in val_FileNames.tolist()]
    test_FileNames = [src + '/' + name for name in test_FileNames.tolist()]

    # Print
    print('Número Total de Imagens: ', len(allFileNames))
    print('Imagens de Treino: ', len(train_FileNames))
    print('Imagens de Validação: ', len(val_FileNames))
    print('Imagens de Teste: ', len(test_FileNames))
    
    # Copia as 
    try:
        for name in train_FileNames:
            shutil.copy(name, "base_data/melspectrogram/model_data/treino/" + nome_classe)
    except:
        pass
    try:
        for name in val_FileNames:
            shutil.copy(name, "base_data/melspectrogram/model_data/val/" + nome_classe)
    except:
        pass
    try:    
        for name in test_FileNames:
            shutil.copy(name, "base_data/melspectrogram/model_data/teste/" + nome_classe)    
    except:
        pass
```
