```python
from google.cloud import storage

# Cria um cliente de storage
storage_client = storage.Client()
```

```python
buckets = storage_client.list_buckets()

print("Buckets:")
for bucket in buckets:
    print(bucket.name)
```

```python
import pandas as pd

bucket_name = 'projeto_musical'
csv_file_name = 'labels.csv'
local_file_path = '/content/' + csv_file_name

# Cria um objeto bucket
bucket = storage_client.get_bucket(bucket_name)

# Cria um objeto blob
blob = bucket.blob(csv_file_name)

# Baixa o arquivo para o ambiente local
blob.download_to_filename(local_file_path)

# Agora o arquivo CSV está salvo localmente e você pode lê-lo usando pandas
df = pd.read_csv(local_file_path)
print(df.head())
```

```python
bucket_name = 'projeto_musical'
rar_file_name = 'imagens_audio.rar'
local_rar_path = '/content/' + rar_file_name

# Define o bucket e o blob
bucket = storage_client.get_bucket(bucket_name)
blob = bucket.blob(rar_file_name)

# Baixa o arquivo RAR para o ambiente local do Colab
blob.download_to_filename(local_rar_path)

# Descompacta o arquivo RAR
!mkdir /content/unpacked
!unrar x {local_rar_path} /content/unpacked/

# Agora você pode listar os arquivos descompactados e carregá-los de volta para o GCS


# Lista os arquivos descompactados


print("Todos os arquivos foram carregados de volta para o GCS.")
```

```python
destination_folder = '/content/unpacked/imagens_audio/chroma_stft/'
blobs = list(storage_client.list_blobs(bucket, prefix=destination_folder))
existing_files = {blob.name.replace(destination_folder, '') for blob in blobs}
```

```python
unpacked_files_path = '/content/unpacked/imagens_audio/chroma_stft'
unpacked_files = os.listdir(unpacked_files_path)
print("Arquivos descompactados:")
for file_name in unpacked_files:
    print(file_name)

    # Define o caminho local do arquivo
    local_file_path = os.path.join('/content/unpacked/imagens_audio/chroma_stft/', file_name)

    # Cria um blob no GCS para cada arquivo descompactado
    new_blob = bucket.blob('base_data/chroma_stft/' + file_name)  # 'pasta_destino/' é onde os arquivos serão armazenados no GCS

    # Carrega o arquivo para o GCS
    new_blob.upload_from_filename(local_file_path)
```
