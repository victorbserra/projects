```python
import pandas as pd
import requests
from pytrends.request       import TrendReq
from google.cloud           import bigquery
from google.oauth2          import service_account
from google.cloud.bigquery  import Client
```

```python
credentials = service_account.Credentials.from_service_account_file('methodius-advisory-86691f61031e.json')
client = bigquery.Client(credentials=credentials, project='methodius-advisory')
```

```python
def get_data(sql_file, var):  
    if var == True:
        dados = client.query(sql_file)
        result = dados.result()
        df = result.to_dataframe()
        return df
    else:
        with open(sql_file, 'r') as f:
            sql = f.read()
        dados = client.query(sql)
        result = dados.result()
        df = result.to_dataframe()
        return df
```

```python
# Configura a conexão com o Google Trends
pytrends = TrendReq(geo='IT', timeout=(10,25))

# Lista de termos de busca relevantes para imóveis em Milão
kw_list = ['immobiliare.it', 'idealista.it', 'trovacasa.it','wikicasa.it']

# Cria a requisição para obter dados dos últimos 12 meses
pytrends.build_payload(kw_list, timeframe='today 12-m')

# Obtemos o interesse ao longo do tempo
data = pytrends.interest_over_time()
# print(data.head())
```

```python
from googleapiclient.discovery import build
```

```python
service = build('searchconsole', 'v1', credentials=credentials)
```

```python
site_url = 'https://www.immobiliare.it/'
```

```python
request = {
    'startDate': '2024-09-01',  # Data inicial
    'endDate': '2024-09-30',    # Data final
    'dimensions': ['query'],    # Filtrando por palavras-chave
    'rowLimit': 10              # Limite de 10 resultados
}

# Fazendo a chamada à API
response = service.searchanalytics().query(siteUrl=site_url, body=request).execute()

# Exibindo os resultados
for row in response.get('rows', []):
    keyword = row['keys'][0]  # Palavra-chave
    clicks = row['clicks']    # Número de cliques
    impressions = row['impressions']  # Impressões
    position = row['position']  # Posição média

    print(f"Palavra-chave: {keyword}, Cliques: {clicks}, Impressões: {impressions}, Posição média: {position}")
```
